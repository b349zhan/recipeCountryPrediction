```python
## Data Loading
import pandas as pd

## Data Manipulation
import numpy as np

## Plotting
import matplotlib.pyplot as plt
import seaborn as sns

## NLP Vectorizer
from sklearn.feature_extraction.text import CountVectorizer

## Models
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import confusion_matrix
```

# 1. Load Data

```python
## Read Training and Test Data From json files in current directory
trainData = pd.read_json("../data/train.json")
testData = pd.read_json("../data/test.json")
print(trainData.shape, testData.shape)
```

# 2. Exploratory Analysis

```python
trainData.head()
```

## 2.1 Country Distribution in Training Data

```python
trainData.cuisine.value_counts().plot.barh()
plt.xlabel("Count")
plt.ylabel("Country")
plt.title("Recipe Count over Countries")
```

## 2.2 Distinct Ingredient Count over Country

```python
countries = np.unique(trainData.cuisine)
ingredientOverCountry = {}
for country in countries:
    countryData = trainData[trainData.cuisine == country]
    if country not in ingredientOverCountry:
        ingredientOverCountry[country] = set()
    for idx in range(countryData.shape[0]):
        ingredientOverCountry[country] = ingredientOverCountry[country].union(\
                                                set(list(countryData.ingredients.iloc[idx])))
counts = [len(ingredientOverCountry[country]) for country in countries]
pairs = [(countries[i], counts[i]) for i in range(len(countries))]
pairs = sorted(pairs, key = lambda i:i[1])
plt.barh([pairs[i][0] for i in range(len(countries))],[pairs[i][1] for i in range(len(countries))] )
plt.title('Distinct Ingredient Count over Countries')
plt.ylabel('Country')
plt.xlabel('Count')
plt.show()
```

```python
ingredients = set()
for i in ingredientOverCountry:
    ingredients = ingredients.union(ingredientOverCountry[i])
print("There are a total of "+str(len(ingredients))+" unique ingredients")
```

# 3. Vectorize inputs

```python
trainData["ingredientStr"] = [" ".join(trainData.ingredients.iloc[i]) for i in range(trainData.shape[0])]
trainData.head()
```

```python
## Define our vectorizer
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer, PorterStemmer
import re
import nltk
def my_tokenizer(text):
    text = text.lower()
    text = re.sub("\d+", " ", text)
    tokenizer = RegexpTokenizer(r'\w+')
    token = [word for word in tokenizer.tokenize(text) if len(word) > 2]
    return token
binary = CountVectorizer(tokenizer = my_tokenizer, binary = True)
X = binary.fit_transform(trainData.ingredientStr)
y = trainData.cuisine
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, stratify = y, 
                                                    shuffle = True, random_state = 42)
```

# 4. Soft Margin Classifier

```python
testAccs = []
trainAccs = []
m = 10
for c in range(1,m):
    print(c,end="")
    svc = SVC(C = c)
    svc.fit(X_train, y_train)
    trainAccs.append((svc.predict(X_train)==y_train).mean())
    testAccuracy = (svc.predict(X_test)==y_test).mean()
    testAccs.append(testAccuracy)
plt.plot(list(range(1,m)),trainAccs, label="Training Accuracy")
plt.plot(list(range(1,m)), testAccs, label="Test Accuracy")
plt.xlabel("C")
plt.ylabel("Accuracy")
plt.title("Train/Test Accuracy vs C")
plt.legend()
```

```python
svc = SVC(C = c)
svc.fit(X_train, y_train)
svctrainAccuracy = (svc.predict(X_train)==y_train).mean()
svcPred = svc.predict(X_test)
svctestAccuracy = (svcPred==y_test).mean()
```

# 5. Logistic Regression

```python
testAccs = []
trainAccs = []
m = 10
for c in range(1,m):
    print(c,end="")
    log = LogisticRegression(penalty="l2",C = c, multi_class = "multinomial", max_iter = 1000)
    log.fit(X_train, y_train)
    trainAccs.append((log.predict(X_train)==y_train).mean())
    testAccuracy = (log.predict(X_test)==y_test).mean()
    testAccs.append(testAccuracy)
plt.plot(list(range(1,m)),trainAccs, label="Training Accuracy")
plt.plot(list(range(1,m)), testAccs, label="Test Accuracy")
plt.xlabel("C")
plt.ylabel("Accuracy")
plt.title("Train/Test Accuracy vs C")
plt.legend()
```

```python
log = LogisticRegression(penalty="l2", multi_class = "multinomial", max_iter = 1000)
log.fit(X_train, y_train)
logtrainAccuracy = (log.predict(X_train)==y_train).mean()
logtestAccuracy = (log.predict(X_test)==y_test).mean()
```

# 6. Random Forest / Extra Trees (Serious Overfitted)

```python
testAccs = []
trainAccs = []
for c in [1,4,8,10]:
    print(c,end="")
    rf = RandomForestClassifier(n_estimators = c*100)
    rf.fit(X_train, y_train)
    trainAccs.append((rf.predict(X_train)==y_train).mean())
    testAccuracy = (rf.predict(X_test)==y_test).mean()
    print(testAccuracy)
    testAccs.append(testAccuracy)
plt.plot([100,400,800,1000],trainAccs, label="Training Accuracy")
plt.plot([100,400,800,1000], testAccs, label="Test Accuracy")
plt.xlabel("Number of Estimators")
plt.ylabel("Accuracy")
plt.title("Train/Test Accuracy vs Number of Estimators")
plt.legend()
```

```python
testAccs = []
trainAccs = []
for c in [1,4,8,10]:
    print(c,end="")
    extra = ExtraTreesClassifier(n_estimators = c*100)
    extra.fit(X_train, y_train)
    trainAccs.append((extra.predict(X_train)==y_train).mean())
    testAccuracy = (extra.predict(X_test)==y_test).mean()
    testAccs.append(testAccuracy)
plt.plot([100,400,800,1000],trainAccs, label="Training Accuracy")
plt.plot([100,400,800,1000], testAccs, label="Test Accuracy")
plt.xlabel("Number of Estimators")
plt.ylabel("Accuracy")
plt.title("Train/Test Accuracy vs C")
plt.legend()
```

```python
confusion = confusion_matrix(y_test, svcPred)
plt.figure(figsize = (20,10))
confusion = confusion/np.sum(confusion, axis = 1)
confusionDf = pd.DataFrame(confusion, index=[i for i in countries], columns=[
                         i for i in countries])
sns.heatmap(round(confusionDf, 2), annot=True, annot_kws={"size": 14})
plt.title("Confusion Matrix for SVC", fontsize=30)
```

# Conclusion 

From the above, we can see that the soft margin classifier performs better than logistic regression and random forest/extra trees. It is interesting that Random Forest and Extra Trees Classifier achieves almost 100% accuracy, which is a huge indicator of overfitting. These two classifiers are bagging ensemble classifiers, which should help with dealing overfitting problems. The upper bounad of loss of these two classifiers are positively correlated to the randomness of the model. 

We can still see that the improvement on the randomness brought up the accuracy a little bit by observing the accuracy between Random Forest and Extra Trees.
